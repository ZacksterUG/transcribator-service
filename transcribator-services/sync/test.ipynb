```python
import nats
```

```python
MAIN_TOPIC = 'transcriber.sync'
nats_connection = await nats.connect(servers='nats://localhost:4222')
```

```python
import json

inbox = nats_connection.new_inbox()
res = await nats_connection.publish(f'{MAIN_TOPIC}.init', json.dumps({'job_id': 'test-123'}).encode(), reply=inbox)
```
